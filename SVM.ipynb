```python
# Importing necessary libraries
```

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn   # For machine learning tools
```

```python
# Loading the breast cancer dataset from sklearn
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
print(cancer['data'])
```

```python
print(cancer['feature_names'])
```

```python
# Creating a DataFrame from the raw data to visualize it better
```

```python
df_feat = pd.DataFrame(cancer['data'],columns = cancer['feature_names'])
df_feat.head(1)
```

```python
# Displaying basic info about the DataFrame
```

```python
df_feat.info()
```

```python
# Checking the target labels (Malignant/Benign)
```

```python
cancer['target_names']
```

```python
# Displaying the actual target values (0 or 1)
```

```python
cancer['target']
```

```python
# Creating a DataFrame for the target column
```

```python
df_target = pd.DataFrame(cancer['target'],columns = ['Cancer'])
```

```python
# Combining the features and target into one DataFrame for easier analysis
```

```python
df = pd.concat([df_feat, df_target], axis = 1)
```

```python
df.head() # Displaying the first few rows
```

```python
# Visualizing relationships between two features with a scatter plot
```

```python
sns.scatterplot(x='mean radius', y='mean texture', hue='Cancer', data=df)
```

```python
# Plotting a count of cancer cases (Malignant vs. Benign)
```

```python
sns.countplot(x = 'Cancer', data = df,)
```

```python
# Splitting the data into training and test sets (70% train, 30% test)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_feat, np.ravel(df_target), test_size = 0.30, random_state = 101)
```

```python
# Training a Support Vector Classifier (SVC)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(x_train,y_train) # Fitting the model on training data
```

```python
# Making predictions on the test set
```

```python
predictions = model.predict(x_test)
```

```python
# Evaluating the model using confusion matrix and classification report
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print (confusion_matrix(y_test,predictions))
```

```python
print (classification_report(y_test,predictions))
```

```python
# Using GridSearchCV to find the best hyperparameters for the SVM model
```

```python
param_grid = {'C':[0.1,1, 10, 100, 1000], 'gamma':[1,0.1,0.01, 0.001, 0.0001], 'kernel':['rbf']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=3) # Running grid search
```

```python
grid.fit(x_train,y_train) # Fitting the model with the best hyperparameters
```

```python
# Making predictions using the optimized model
```

```python
grid_predictions = grid.predict(x_test)
```

```python
# Evaluating the optimized model's performance
```

```python
print(confusion_matrix(y_test,grid_predictions))
```

```python
print(classification_report(y_test,grid_predictions))
```
